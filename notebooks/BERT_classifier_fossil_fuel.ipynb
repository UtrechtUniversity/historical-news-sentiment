```python
from pathlib import Path
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
```

```python
def confusion_classification_report(prediction_fp, method_title):
    print("\n" + "=" * len(method_title))
    print(method_title.upper())
    print("=" * len(method_title))
    
    df = pd.read_csv(prediction_fp)
    y_true = df['true_labels']
    y_pred = df['predictions']
    
    cm = confusion_matrix(y_true, y_pred)
    print("\nConfusion Matrix:\n", cm)
    
    report = classification_report(y_true, y_pred, digits=3)
    print("\nClassification Report:\n", report)
```

```python
prediction_fp="../../output/fossil_fuel/predict/predictions_bertje2.csv"
method_title="BERTje Sentiment Model: Confusion Matrix and Classification Report (Binary Classification)"
confusion_classification_report(prediction_fp, method_title)

prediction_fp="../../output/fossil_fuel/predict/predictions_gysbert2.csv"
method_title="GysBERT Sentiment Model: Confusion Matrix and Classification Report (Binary Classification)"
confusion_classification_report(prediction_fp, method_title)

prediction_fp="../../output/fossil_fuel/predict/predictions_robbert2.csv"
method_title="RobBERT Sentiment Model: Confusion Matrix and Classification Report (Binary Classification)"
confusion_classification_report(prediction_fp, method_title)
```

```python

bertje_prediction_fp="../../output/fossil_fuel/predict/predictions_bertje2.csv"
gys_prediction_fp="../../output/fossil_fuel/predict/predictions_gysbert2.csv"
robbert_prediction_fp="../../output/fossil_fuel/predict/predictions_robbert2.csv"

bertje_df = pd.read_csv(bertje_prediction_fp)
y_true = bertje_df['true_labels']
bertje_y_pred = bertje_df['predictions']

gys_df = pd.read_csv(gys_prediction_fp)
gys_y_pred = gys_df['predictions']

robbert_df = pd.read_csv(robbert_prediction_fp)
robbert_y_pred = robbert_df['predictions']



predictions = {
    'BERTJe': bertje_y_pred,
    'GysBERT': gys_y_pred,
    'RobBERT': robbert_y_pred,
}

plt.figure(figsize=(8, 6))

for method, y_scores in predictions.items():
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f'{method} (AUC = {roc_auc:.3f})')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Random')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison - Binary Classification')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
prediction_fp="../../output/fossil_fuel/predict/predictions_bertje3.csv"
method_title="BERTje Sentiment Model: Confusion Matrix and Classification Report (multiclass Classification)"
confusion_classification_report(prediction_fp, method_title)

prediction_fp="../../output/fossil_fuel/predict/predictions_gysbert3.csv"
method_title="GysBERT Sentiment Model: Confusion Matrix and Classification Report (multiclass Classification)"
confusion_classification_report(prediction_fp, method_title)

prediction_fp="../../output/fossil_fuel/predict/predictions_robbert3.csv"
method_title="RobBERT Sentiment Model: Confusion Matrix and Classification Report (multiclass Classification)"
confusion_classification_report(prediction_fp, method_title)
```
